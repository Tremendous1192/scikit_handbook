```python
# ライブラリのインポート
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions
```

```python
# XORのデータを作成する(x=正、y=正)=0,(x=正、y=負)=1
X_xor = np.array([[1.0,0],[-1.0,0],[0,-2.0],[0,2]])
y_xor = np.array([0,0,1,1])                 

df_X = pd.DataFrame(X_xor)
df_X.columns = [u'x1', u'x2']

df_y = pd.DataFrame(y_xor)
df_y.columns = [u'XOR']

# df_X+df_yの先頭5行の表示
pd.concat([df_X, df_y], axis=1).head()
```

```python
# データのプロット
plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1], c='b', marker='x', label='1')
plt.scatter(X_xor[y_xor==0, 0], X_xor[y_xor==0, 1], c='r', marker='s', label='0')
plt.legend(loc='best')
plt.show
```

```python
# 訓練データとテストデータに分割
X_train, y_train = X_xor, y_xor
print('X_trainの形状：', X_train.shape)
print('y_trainの形状：', y_train.shape)
```

```python
# SVCのモデルを作成
model = SVC(kernel='rbf', gamma=700 , C=10,random_state=0)

# モデルの訓練
model.fit(X_train, y_train)
```

```python
# 訓練データのプロット
plt.figure(figsize=(8,4)) #プロットのサイズ指定
plot_decision_regions(X_train, y_train, model)
```
